### [실습 1] 중심 경향 및 산포 측도 계산


아래 배열에서 평균, 중앙값, 최빈값, 분산, 표준편차, 범위를 계산하세요.


```python
	arr = np.array([10, 20, 30, 30, 40, 50, 60])
```

````python
import numpy as np
arr = np.array([10, 20, 30, 30, 40, 50, 60])
print(np.mean(arr))
print(np.median(arr))
# arr 최빈값


# 최빈값: 가장 자주 등장하는 값
# -> np.unique(ndarray, return_counts=True)
# return_counts=True 옵션: 중복을 제거하면서 같은 값이 몇 개 있는지 카운트해서 반환
# print(np.unique(scores, return_counts=True))
values, counts = np.unique(arr, return_counts=True)
max_count = np.max(counts)
# 중복 제거된 배열에서 최빈값을 가진 수만 찾기
vals = values[counts == max_count]
print("최빈값: ", vals)


print(np.var(arr))
print(np.std(arr))
````
