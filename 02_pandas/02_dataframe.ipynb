# DataFrame

```python
import numpy as np
import pandas as pd
```

## DataFrame 생성

```python
# dict -> df 생성
data = {"one": [1.0, 2.0, 3.0], "two": [10, 20, 30], "three": 99}

# dict의 각 item(key:value)이 하나의 Series로 처리됨
df = pd.DataFrame(data)
print(df["one"])    # "one" 열의 Series 선택
df  # broadcasting

```

```python
# 여러 개의 dict -> dataframe
data: list[dict[str, int]] = [{
    "a": 10,
    "b": 20,
}, {
    "a": 30,
    "b": 40,
    "c": 50,
}]

# 하나의 dict -> 하나의 레코드, key는 컬럼으로 처리
df = pd.DataFrame(data)
df  # NaN == 결측치
```

```python
# ndarray -> df
arr = np.arange(9).reshape(3, 3)
df = pd.DataFrame(arr, columns = ("a", "b", "c"), index=['1일', '2일', '3일'])
df
```

```python
# ser -> df
name_ser = pd.Series(['홍길동', '이순신', '강감찬'])
age_ser = pd.Series([20, 25, 30])
df = pd.DataFrame({"name": name_ser, "age": age_ser})
df
```

```python
# tuple(record) -> dataframe
# 하나의 객체

data = [
    ('홍길동', 20),
    ('이순신', 30),
    ('논개', 43),
]
df = pd.DataFrame.from_records(data, columns=['이름', '나이'])
df
```

## 조회
- 행열조회
    - df.iloc[인덱스]
    - df.loc[라벨]
- 열조회
    - df[컬럼명]

### iloc (인덱스 위치)

```python
dicts = [
    {'name' : '홍길동', 'job' : '강사', 'age' : 33},
    {'name' : '신사임당', 'age' : 44, 'job' : '영업사원'},
    {'name' : '이순신', 'age' : 38, 'job' : '장군'},
    {'name' : '논개', 'age' : 25, 'job' : '개발'},
    {'name' : '강감찬', 'age' : 47, 'job' : '교수'},
]
df = pd.DataFrame(dicts)
df
```

```python
# 홍길동 행 조회
print(df.iloc[0])
print(type(df.iloc[0])) # type == Series, df에서 한 행 또는 한 열을 Series 타입이다.
print(df.iloc[0].index)
print(df.iloc[0].name)  # fd의 인덱스  == ser의 이름
```

```python
# 슬라이싱 지원
df2 = df.iloc[0:2]  # 0, 1행 반환  (2차원) -> df 반환
print(type(df2))
print(df2)
```

```python
# fancy indexing 지원
df.iloc[[0, 2]] # 0번 행, 2번 행ㅠ
```

```python
# 단일 값 조회
print(df.iloc[0, 1])    # 인덱스 기반이므로 ['job'] X

# 여러 행 조회
df.iloc[:2, 1]
```

```python
df2 = df.iloc[:2, :2]
print(type(df2))
df2
```

### loc
인데스라벨을 통해 조회

```python
# index명 지정 == 라벨링(labeling)
df.index = ['client1', 'client2', 'client3', 'client4', 'client5']
df
```

```python
print(type(df.loc['client4']))
df.loc['client4']
```

```python
df.iloc[3]

# df.loc[3]   # 그냥 loc는 인덱스
```

# loc 슬라이싱은 end 값을 포함한다!

```python
# loc를 이용한 컬럼 단일 값 조회
df.loc['client3', 'job']

# df.iloc[0, 1]

# loc를 이용한 슬라이싱
print(df.loc['client2': 'client4']) # client4 포함
```

```python
# loc를 이용한 슬라이싱 + 컬럼 선택
print(df.loc['client2':'client4', 'job'])

print(df.loc['client2':'client4', 'job':])
```

### 컬럼라벨

```python
# df['컬럼명'] : 해당 컬럼 시리즈 반환
df['job']
print(type(df['name']))
df['name']
```

```python
# 컬럼 라벨 fancy indexing
print(type(df[['name', 'job']]))
df[['name', 'job']]
```

```python
# loc # 레이블 이름으로 선택하는 것
# loc + 컴럼 라벨
print(df['name'] == '이순신')
print('-----------------------------')
print(df.loc[df['name'] == '이순신'])
```

```python
# boolean indexing (비교 연산)
print(df['age'] > 30)
df[df['age'] > 30]
```

```python
# 30세 미만, 개발 직종을 갖는 행 조회
df.loc[(df['age'] < 30) & (df['job'] == '개발')]
```

```python
# - 이름에 '신'이 들어가거나 나이가 40대이상 조회 (OR 연산)
df[(df['age'] >= 40) | (df['name'].str.contains('신'))]
```
