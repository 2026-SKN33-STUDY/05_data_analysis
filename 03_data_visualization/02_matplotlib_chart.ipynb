# 차트유형별 데이터 분석
데이터의 종류에 따라 데이터 분석에 적절한 시각화차트를 선택할수 있어야 한다.

**이산값 | 연속값 구분**
- **이산값:**
    - 값이 띄엄띄엄 떨어져 있다는 뜻으로 값이 이어져 있지 않고, 정해진 값 중에 하나만 가질 수 있는 값이다.
    - 분류 문제에서 레이블은 보통 이산값을 갖는다.
        - 예를 들어, 이메일이 스팸인지 아닌지를 분류하는 문제에서는 두 개의 이산값(스팸, 스팸 아님)이 사용된다.
        - 회원등급, 성별 등 카테고리(라벨)로 정의되는 값을 의미함
- **연속값:**
    - 값이 끊어지지 않고 이어질 수 있는 값이다.
    - 회귀 문제에서는 예측해야 하는 값이 연속적이다.
    - 예를 들어, 내일의 온도를 예측하는 문제에서는 예측된 온도가 연속적인 값
    - 주식 가격, 환율, 이자율 등의 시계열 데이터는 모두 연속값

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fontTools.diff import color
````

````python
# Matplotlib은 기본 폰트가 한글을 지원하지 않아 경고가 날 수 있습니다.
# 운영체제별 후보 폰트 중 실제 설치된 폰트를 찾아 그래프 제목/축 라벨이 깨지지 않게 설정합니다.

import platform
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# Matplotlib 기본 폰트(DejaVu Sans)는 한글 글리프가 없어 경고가 날 수 있습니다.
# 운영체제별로 사용 가능한 한글 폰트를 찾아 설정합니다.
font_candidates = {
    "Darwin": ["AppleGothic", "NanumGothic", "Noto Sans CJK KR", "Noto Sans KR"],
    "Windows": ["Malgun Gothic", "NanumGothic", "Noto Sans CJK KR", "Noto Sans KR"],
    "Linux": ["NanumGothic", "Noto Sans CJK KR", "Noto Sans KR"],
}

available_fonts = {font.name for font in fm.fontManager.ttflist}
system_name = platform.system()
selected_font = None

for font_name in font_candidates.get(system_name, []):
    if font_name in available_fonts:
        selected_font = font_name
        break

if selected_font:
    plt.rcParams["font.family"] = selected_font
    print("설정된 한글 폰트:", selected_font)
else:
    print("사용 가능한 한글 폰트를 찾지 못했습니다. NanumGothic 또는 Noto Sans KR 설치가 필요할 수 있습니다.")

# 마이너스 기호가 깨지는 문제 방지
plt.rcParams["axes.unicode_minus"] = False
````

## Histogram
연속형데이터를 일정간격으로 나누고, 구간별 데이터 빈도수를 시각화

````python
scores = [88, 90, 77, 65, 80, 85, 92, 78, 75, 82, 94, 91, 60, 73, 86, 82, 93, 72, 75, 68]

# plt.hist(): 히스토그램
# - count, bins, container 반환
# - count: 각 구간 데이터의 개수
# - bins: 구간 경계(막대 개수)
# - container: 막대 객체

count, bins, container = plt.hist(
    scores,
    bins=8, # 막대 개수
    rwidth=0.9, # 막대 너비
    edgecolor="black",  # 테두리 색상
)

plt.xlabel("Scores")
plt.ylabel("Count")
plt.show()


# 그래프의 자세한 정보 출력
print("count:", count)
print("bins:", bins)
print("container:", container)
print('범위:')
for i in range(len(count)):
    print(f'{bins[i]} ~ {bins[i+1]}: {int(count[i])}개')
````

````python
# scores 재사용

bins = np.linspace(0, 100, 11)  # 0~ 100 균등한 간격으로 숫자 생성
print(bins)

# x축 범위를 별도로 지정한 히스토그램 생성
plt.hist(scores, bins=bins, rwidth=0.9, edgecolor="black")


plt.xlabel("Scores")
plt.ylabel("Count")
plt.show()
````

````python
# 누적 히스토그램  # 겹쳐있는 히스토그램
# 지역별 점심 식사 비용 분포를 누적 히스토그램으로 비교
np.random.seed(100) # 똑같은 값이 나올 수 있게
region1 = np.random.normal(8000, 2000, 100)    # 평균 8000원, 표준편차, 100 개
region2 = np.random.normal(10000, 3000, 100)
region3 = np.random.normal(12000, 4000, 100)


plt.hist(
    [region1, region2, region3],
    label=["Region 1", "Region 2", "Region 3"],
    stacked=True,   # 누적
    rwidth=0.8,
    edgecolor="black",
)

plt.xlabel("Cost")
plt.ylabel("Count")

plt.legend()
plt.show()
````

## Bar Plot
막대그래프

이산형데이터에 대한 수치값(빈도수, 수치)을 막대의 길이로 표현하는 그래프

````python
# 월별 판매 금액을 세로 막대 그래프로 출력
months = ['1월', '2월', '3월', '4월', '5월', '6월']

# 판매 금액 정규 분포 난수로 6개 생성 (int)
np.random.seed(1)
sales = np.random.normal(3_000_000, 500_000, 6).astype(int)

plt.bar(months, sales , color="skyblue", edgecolor="black")

plt.title("월별 판매 금액")
plt.xlabel("월")
plt.ylabel("판매금액(백만)")

# plt.text(x, y): 지정된 좌표에 문자열 출력
for i, money in enumerate(sales):
    plt.text(i,
             money + 50000,
             str(money),
             fontsize=8,
             ha="center"   # horizonal align 정렬
             )

plt.show()
````

````python
# 가로 막대 그래프
# 직업별 연봉을 정렬한 뒤 가로 막대 그래프로 출력
job_df = pd.DataFrame([
    ('개발자', 8000),
    ('의사', 10000),
    ('교사', 7000),
    ('작가', 6000),
    ('변호사', 9000),
], columns=['직업', '연봉']).sort_values('연봉', ascending=True)

job_df

# plt.barh(): 가로 막대 그래프 horizontal
plt.barh(job_df['직업'], job_df['연봉'])
plt.ylabel('직업')
plt.xlabel('연봉')

# plt.text(x, y): 지정된 좌표에 문자열 출력
for i, data in enumerate(job_df['연봉']):
    plt.text(data-700, i,
             str(data),
             fontsize=12,
             va="center",   # vertical align 정렬
             ha="center"    # horizontal align 정렬
             )




plt.show()
````

## Box Plot
연속형데이터의 분포와 동시에 통계지표(사분위수-최소값/최대값/중위값)를 시각화.

### 이상치 표시하기
Box Plot은 데이터의 이상치(outlier)를 자동으로 감지하여 표시한다.
이상치는 수염(whisker) 바깥쪽에 있는 값으로, 박스 바깥에 작은 점으로 표시된다.

이상치(Outlier)는 보통 데이터의 분포에서 크게 벗어난 값들을 말하며, Boxplot에서 이상치를 판별하는 기준은 일반적으로 **IQR(Interquartile Range, 사분위수 범위)**를 기반으로 한다.

**Boxplot 구성 요소 요약:**
```txt
      ○  이상치 (outlier)

      |------ whisker top (최댓값 또는 이상치 전까지)
      |       (이상치 있을 경우: 이상치 직전까지)
      |
      |
      |
 Q3 ──┐────┐  ← 3사분위수 (75%)
      │    │
      │ Box│
      │    │
 Q2 ──┼────┘  ← 중앙값 (50%)
      │
      │
 Q1 ──┘         ← 1사분위수 (25%)
      |
      |------ whisker bottom (최솟값 또는 이상치 전까지)
             (이상치 있을 경우: 이상치 직전까지)
```

**핵심 설명:**
* Q1 (1사분위수): 하위 25% 지점
* Q2 (중앙값, Median): 50% 지점
* Q3 (3사분위수): 상위 75% 지점
* IQR (Interquartile Range): Q3 - Q1
* Whisker (수염):
    * 이상치가 없을 경우 최솟값과 최댓값,
    * 이상치가 있을 경우 Q1 - 1.5×IQR ~ Q3 + 1.5×IQR 범위 내의 값


**이상치 판별**
위에서 구한 하한선보다 작은 값이나 상한선보다 큰 값들을 이상치로 간주한다.


**이상치가 있는 경우:**
* 맨 아랫값 (whisker bottom): Q1 - 1.5×IQR 이상인 값 중 최소값
* 맨 윗값 (whisker top): Q3 + 1.5×IQR 이하인 값 중 최댓값
* 그보다 바깥에 있는 값들은 ○ 같은 점으로 **이상치(outlier)**로 표시됨


````python
# 상품별 가격 데이터에서 이상치 후보를 확인 (box plot)
data_with_outliers = [
    [100, 120, 130, 110, 115, 105, 125, 200],   # 이상치 200 포함
    [150, 160, 170, 155, 165, 158, 162],
    [80, 85, 90, 75, 92, 87, 300],              # 이상치 300 포함
    [200, 220, 210, 205, 215, 230, 50],         # 이상치 50 포함
]

plt.boxplot(
    data_with_outliers,
    patch_artist=True,
    boxprops=dict(facecolor='skyblue')
)


plt.show()
````

````python
# 이상치 교정
# IQR 방식으로 박스플롯 이상값 후보 계산 과정

# 첫 번째 상품 데이터를 배열로 변환해 IQR 계산 대상 지정
data = np.array(data_with_outliers[0])
data

# q3: 데이터의 75% 지점
q3 = np.quantile(data, 0.75)

# q1: 데이터의 25% 지점
q1 = np.quantile(data, 0.25)

# IQR = Q3 - Q1, 가운데 50% 데이터 범위
iqr = q3 - q1

print('q1 =', q1)
print('q3 =', q3)
print('iqr =', iqr)

# lower_bound: IQR 기준 이상값 하한
lower_bound = q1 - 1.5 * iqr

# upper_bound: IQR 기준 이상값 상한
upper_bound = q3 + 1.5 * iqr
print('lower bound =', lower_bound)
print('upper bound =', upper_bound)

# 하한보다 작거나 상한보다 큰 값만 이상값 후보로 추출
outliers = data[(data < lower_bound) | (data > upper_bound)]
outliers# IQR 방식으로 박스플롯 이상값 후보 계산 과정

# 첫 번째 상품 데이터를 배열로 변환해 IQR 계산 대상 지정
data = np.array(data_with_outliers[0])
data

# q3: 데이터의 75% 지점
q3 = np.quantile(data, 0.75)

# q1: 데이터의 25% 지점
q1 = np.quantile(data, 0.25)

# IQR = Q3 - Q1, 가운데 50% 데이터 범위
iqr = q3 - q1

print('q1 =', q1)
print('q3 =', q3)
print('iqr =', iqr)

# lower_bound: IQR 기준 이상값 하한
lower_bound = q1 - 1.5 * iqr

# upper_bound: IQR 기준 이상값 상한
upper_bound = q3 + 1.5 * iqr
print('lower bound =', lower_bound)
print('upper bound =', upper_bound)

# 하한보다 작거나 상한보다 큰 값만 이상값 후보로 추출
outliers = data[(data < lower_bound) | (data > upper_bound)]
outliers

# 이상치를 경계값으로 대치
data = data.clip(lower_bound, upper_bound)
data

plt.boxplot(data)
plt.show()
````
